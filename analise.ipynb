```python
import pandas as pd
```

```python
df = pd.read_csv('furtoSimples3.csv', encoding='utf-8', delimiter=",")
```

```python
df.head()
```

```python
# Selecionar a primeira coluna
numSentenca = df.iloc[:, 0]

# Selecionar as últimas 6 colunas
ultimas_seis = df.iloc[:, -6:]
```

```python
ultimas_seis.head()
```

```python
tabela_df = pd.concat([numSentenca, ultimas_seis], axis=1)
```

```python
tabela_df.head()
```

```python
for i in range(1, 7):
    # Pega o índice da coluna contando de trás pra frente
    col_idx = -i
    
    # Pega o nome da coluna
    col_name = tabela_df.columns[col_idx]
    
    # Passo 1: substitui vírgulas por pontos
    coluna_convertida = tabela_df.iloc[:, col_idx].astype(str).str.replace(',', '.', regex=False)
    
    # Passo 2: converte para float (com erros virando NaN)
    coluna_convertida = pd.to_numeric(coluna_convertida, errors='coerce')
    
    # Passo 3: substitui a coluna original
    tabela_df[col_name] = coluna_convertida
```

```python
tabela_df.head()
```

```python
linhas = tabela_df[(tabela_df.iloc[:, -2] >= 4) & (tabela_df.iloc[:, -2] <= 60.83)]
```

```python
linhas
```

```python
linhaf = linhas[(linhas.iloc[:, -5] >= 12.17) & (linhas.iloc[:, -5] <= 36.50)]
```

```python
linhaf
```

```python
# Define os novos nomes das colunas
novos_nomes = ['Sentença', 'Pena_base(dias)', 'Pena_base(meses)', 'Pena_base(anos)', 'Pena_definitiva(dias)', 'Pena_definitiva(meses)', 'Pena_definitiva(anos)']

# Substitui os nomes no DataFrame
linhaf.columns = novos_nomes
linhaf.to_csv("furtoSimples_filtrado.csv", index=False, sep=';')
```

Pena-base (dias)

```python

coluna_alvo = linhaf.iloc[:, 1]

# Calcula as estatísticas
media = coluna_alvo.mean()
mediana = coluna_alvo.median()
desvio_padrao = coluna_alvo.std()
valor_minimo = coluna_alvo.min()
valor_maximo = coluna_alvo.max()
coef_variacao = (desvio_padrao / media) * 100  # CV em %

# Exibe os resultados
print(f'Média: {media:.2f}')
print(f'Mediana: {mediana:.2f}')
print(f'Desvio Padrão: {desvio_padrao:.2f}')
print(f'Mínimo: {valor_minimo:.2f}')
print(f'Máximo: {valor_maximo:.2f}')
print(f'Coeficiente de Variação (CV%): {coef_variacao:.2f}%')
```
